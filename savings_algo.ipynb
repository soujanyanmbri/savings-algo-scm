```python
# calculating savings for each route
savings = dict()
for row in pw.index:
    for col in pw.columns:
        if int(col) != int(row):            
            a = max(int(row), int(col))
            b = min(int(row), int(col))
            key = '(' + str(a) + ',' + str(b) + ')'
            savings[key] = nodes['D0'][int(row)] + nodes['D0'][int(col)] - pw[col][row]

# insert savings into pandas dataframe, and sort by descending
sv = pd.DataFrame.from_dict(savings, orient = 'index')
sv.rename(columns = {0:'saving'}, inplace = True)
sv.sort_values(by = 'saving', ascending = False, inplace = True, kind='quicksort') # Sorting the list of savings in descending order
sv.head(15)
```

```python
# convert route string to list format, i.e. str(5, 2) to (5, 2)
def get_node(link):
  link = link[1:]
  link = link[:-1]
  nodes = link.split(',')
  return [int(nodes[0]), int(nodes[1])]

# determine if a node is interior to a route
def interior(node, route):
  try:
    i = route.index(node)
    # adjacent to depot, not interior
    if i == 0 or i == (len(route) - 1):
      label = False
    else:
      label = True
  except:
    label = False  
  return label

# merge two routes with a connection link
def merge(route0, route1, link):
  if route0.index(link[0]) != (len(route0) - 1):
    route0.reverse()
    
  if route1.index(link[1]) != 0:
    route1.reverse()
        
  return route0 + route1  
```

```python
# sum up to obtain the total cans belonging to a route
def sum_cap(route, li = nodes):
  sum_cap = 0
  for node in route:
    sum_cap += li.demand[node]
  return sum_cap
```
